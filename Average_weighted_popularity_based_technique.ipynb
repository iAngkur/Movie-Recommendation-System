```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv(r'C:\Users\Angkur\Downloads\Machine Learning\Recommendation system - Python\My Movie Recommendation System/clean_data.csv')
```

```python
data.head(2)
```

```python
popularity = data.sort_values('popularity', ascending=False)
```

```python
popularity.head(10)
```

```python
popularity.columns
```

```python
pop = popularity[['original_title', 'popularity', 'vote_average', 'vote_count']]
```

```python
pop.head(5) # Top 5 popular movies
```

```python
pop.set_index('original_title')['popularity'].nlargest(10).plot(kind='bar')
plt.title('Most popular movies', weight='bold')
plt.xlabel('Movie title')
plt.ylabel('Score of popularity')
```
